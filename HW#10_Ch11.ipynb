```python
#Question 11.12

def hanoi(n , startrod, endrod, temprod):
    if n == 0:
        return
    hanoi(n-1, startrod, temprod, endrod)
    print("Move ",n," ",startrod,"--->",endrod)
    hanoi(n-1, temprod, endrod, startrod)
          

n = 5
hanoi(n, 'A', 'C', 'B') 
```

```python
#Question 11.11
def display(board):
    row = '|   ' * len(board) + '|'
    hr  = '+---' * len(board) + '+'
    for col in board:
        print(hr)
        print(row[:col*4-3],'Q',row[col*4:])


def issafe(q, board, x=1):
    if not board: return True
    if board[-1] in [q+x,q-x]: return False
    return issafe(q, board[:-1], x+1)

def solve(boardsize, board=[]):
    if len(board) == boardsize:
        display(board)
    else:
        for q in [col for col in range(1,boardsize+1) if col not in board]:
            if issafe(q,board):
                solve(boardsize, board + [q])


print(solve(8))
```

```python
#Question 11.22
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import numpy as np
import random
  
  

def mergesort(A, start, end):
    if end <= start:
        return
  
    mid = start + ((end - start + 1) // 2) - 1

    yield from mergesort(A, start, mid)
    yield from mergesort(A, mid + 1, end)
    yield from merge(A, start, mid, end)
  

def merge(A, start, mid, end):
    merged = []
    leftIdx = start
    rightIdx = mid + 1
  
    while leftIdx <= mid and rightIdx <= end:
        if A[leftIdx] < A[rightIdx]:
            merged.append(A[leftIdx])
            leftIdx += 1
        else:
            merged.append(A[rightIdx])
            rightIdx += 1
  
    while leftIdx <= mid:
        merged.append(A[leftIdx])
        leftIdx += 1
  
    while rightIdx <= end:
        merged.append(A[rightIdx])
        rightIdx += 1
  
    for i in range(len(merged)):
        A[start + i] = merged[i]
        yield A
  

def showGraph():
      
    a = np.array([35, 73, 90, 65, 23, 86, 43, 81, 34, 58])
    random.shuffle(a)


    generator = mergesort(a, 0, len(a)-1)

  
    fig, ax = plt.subplots()
      

    bar_rects = ax.bar(range(len(a)), a, align="edge", 
                       color="green")
      

    ax.set_xlim(0, 11)
    ax.set_ylim(0, 100)

      

    text = ax.text(0.01, 0.95, "", transform=ax.transAxes, 
                   color="red")
    iteration = [0]
  
    def animate(A, rects, iteration):
        for rect, val in zip(rects, A):

            rect.set_height(val)
        iteration[0] += 1
        text.set_text("iterations : {}".format(iteration[0]))

    anim = FuncAnimation(fig, func=animate,
        fargs=(bar_rects, iteration), frames=generator, interval=500,
        repeat=False)
    plt.show()
  
showGraph()
```
